```python
import pandas
import requests
import numpy as np
import io
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.offline as offline
import plotly.graph_objs as go
init_notebook_mode(connected=True)
```

```python
url="https://www.colorado.edu/lab/allen/node/105/attachment"
s=requests.get(url).content
deseqres=pandas.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
deseqres
```

```python
p=deseqres["pvalue"].tolist()
```

```python
print (p)
```

```python
data = [go.Histogram(x=deseqres["pvalue"])]
layout = go.Layout(
    yaxis=dict(title="number of genes",
        autorange=True),xaxis=dict(title="Pvalue",
        autorange=True))
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
notsig=deseqres[deseqres.padj>=0.05]
sig = deseqres[deseqres.padj<0.05]
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=np.log(notsig["baseMean"].tolist()), y=notsig["log2FoldChange"].tolist(),mode='markers',text=notsig["Unnamed: 0"].tolist(), name="not-sig"))
fig.add_trace(go.Scatter(x=np.log(sig["baseMean"].tolist()), y=sig["log2FoldChange"].tolist(),mode='markers',text=sig["Unnamed: 0"].tolist(), name="sig"))
iplot(fig)
```
